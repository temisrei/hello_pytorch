```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

##### y_data = b + w * x_data

```python
w = 1.0 # random value


def forward(x):
    return w * x

def loss(x, y):
    y_head = forward(x)
    return (y_head - y)**2
```

### Compute loss for w

```python
w_list = []
mse_list = [] # mean square error
```

```python
for w in np.arange(0.0, 4.1, 0.1):
    print("w =", w)

    lo_sum = 0 # mean square error summation, initial value
    
    print(" ", " x ", " y ", "y_head", "loss")
    for x_val, y_val in zip(x_data, y_data):
        y_head_val = forward(x_val)
        lo = loss(x_val, y_val) # calculate mean square error
        lo_sum += lo            # mean square error summation
        print(" ", x_val, y_val, y_head_val, lo)
    print("MSE =", lo_sum / len(x_data))
    w_list.append(w)
    mse_list.append(lo_sum / len(x_data))
    print()
```

```python
# mse_list
```

```python
plt.plot(w_list, mse_list)
plt.xlabel('w')
plt.ylabel('Loss')
plt.show()
```
